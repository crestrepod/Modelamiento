## Gráfica

```python
experimental_data_FCS   = pd.read_csv('datos/FCS_experimental.csv',          sep=';', decimal=',')
model_data_FCS          = pd.read_csv('datos/Modelo_FCS.csv',   sep=';', decimal=',')
model_data_FCS_2        = pd.read_csv('datos/Modelo_FCS_2.csv', sep=';', decimal=',')

time_experimental_FCS = experimental_data_FCS['Time_dif_hour']
T_experimental_FCS    = experimental_data_FCS['Temperature(C)']

time_model_FCS = model_data_FCS['Tiempo'] / 3600
T_model_FCS    = model_data_FCS['Temperatura'] -273.15

time_model_FCS_2 = model_data_FCS_2['Tiempo'] / 3600
T_model_FCS_2    = model_data_FCS_2['Temperatura'] -273.15

plt.figure()
plt.title('Extracto STK')
plt.plot(time_experimental_FCS, T_experimental_FCS, label = 'Experimental')
plt.plot(time_model_FCS, T_model_FCS, '--', label = 'Modelo')
plt.plot(time_model_FCS_2, T_model_FCS_2, '--', label = 'Modelo ajustado')
plt.xlabel('Tiempo [h]')
plt.ylabel('Temperatura [°C]')
plt.legend()
plt.show()
```

Este ajuste en el modelo FCS se logra multiplicando por 0.8 el Cp con T > T_cong y por 1.2 el Cp a T < T_cong y ambas densidades
